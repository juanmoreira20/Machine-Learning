```python
import pandas as pd
import numpy as np
import tensorflow as tf

from matplotlib import pyplot as plt
# critério de parada para leitura dos dados
```

```python
#Leitura e inserção do tipo do vinho
db = pd.read_csv(
     '/content/winequality-red.csv',
    sep = ';'


    )
db2 = pd.read_csv(
     '/content/winequality-white.csv',
    sep = ';'


    )
```

```python
Data = db
Data2 = db2
```

```python
#Concatenando os dataframes para juntar o vinho branco e tinto
frames = [Data, Data2]
Data_ = pd.concat(frames)
```

```python
#Separação de dados entre treino e teste
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
X = Data_.iloc[:, 0:11]

#Como os dados das notas vão de 3-9, serão necessários apenas valores de 0-6 para representá-los 
y = np.ravel(Data_.quality-3)
x_train, x_test, y_train, y_test = train_test_split(
		X, y, test_size = 0.2, random_state = 200)
```

```python
y
```

```python
#Pré processamento
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
x_train[0]
```

```python
#import keras 
from keras.models import Sequential
from keras.layers import Dense, Flatten, Dropout

# Iniciando o construtor
model_ = Sequential()

# Adicionando a camada de entrada
model_.add(Dense(11, activation ='relu', input_shape =(11, )))

# Adicionando a camada hidden
model_.add(Dense(130, activation ='relu'))
model_.add(Dense(70, activation ='relu'))
model_.add(Dense(40, activation ='relu'))
model_.add(Dense(13, activation ='relu'))
# Adicionando a saída
model_.add(Dense(7, activation ='softmax'))


model_.output_shape

model_.summary()

model_.get_config()


model_.get_weights()
#Definição do modelo
model_.compile(loss ='sparse_categorical_crossentropy',
optimizer ='adam', metrics =['accuracy'])
```

```python
# Treinamento
model_.fit(x_train, y_train, epochs = 200,
		batch_size = 1, verbose = 1)

# Predição
y_pred = model_.predict(x_test)
```
