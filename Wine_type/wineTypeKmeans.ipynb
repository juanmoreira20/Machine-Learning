```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
# critério de parada para leitura dos dados
stop = 10**15
```

```python
#Leitura e inserção do tipo do vinho
db = pd.read_csv(
     '/content/winequality-red.csv',
    sep = ';'


    )
```

```python
Data = db[:stop]
Data.insert(12,'Type',1)
```

```python
Data
```

```python
#leitura e inserção do tipo do vinho
db2 = pd.read_csv(
     '/content/winequality-white.csv',
    sep = ';'


    )
```

```python
Data2 = db2[:stop]
```

```python
Data2.insert(12,'Type',0)
```

```python
Data2
```

```python
#Concatenando os dataframes para juntar o vinho branco e tinto
frames = [Data, Data2]
Data_ = pd.concat(frames)
Data_
```

```python
#Separação de dados entre treino e teste
from sklearn.model_selection import train_test_split
X = Data_.iloc[:, 0:11]
y = np.ravel(Data_.Type)

x_train, x_test, y_train, y_test = train_test_split(
		X, y, test_size = 0.2, random_state = 200)
```

```python
#Pré processamento
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
X
```

```python
import sklearn
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 2, random_state = 200)
KMmodel = kmeans.fit(x_train)
k = KMmodel.labels_
```

```python
sklearn.metrics.accuracy_score(y_train,k)
```

```python
p = kmeans.predict(x_test)
```

```python
sklearn.metrics.accuracy_score(y_test,p)
```

Rede neural abaixo

```python
#import keras 
from keras.models import Sequential
from keras.layers import Dense

# Iniciando o construtor
model = Sequential()

# Adicionando a camada de entrada
model.add(Dense(12, activation ='relu', input_shape =(11, )))

# Adicionando a camada hidden
model.add(Dense(9, activation ='relu'))

# Adicionando a saída
model.add(Dense(1, activation ='sigmoid'))

model.output_shape

model.summary()

model.get_config()


model.get_weights()
#Definição do modelo
model.compile(loss ='binary_crossentropy',
optimizer ='adam', metrics =['accuracy'])
```

```python
# Treinamento
model.fit(x_train, y_train, epochs = 3,
		batch_size = 1, verbose = 1)

# Predição
y_pred = model.predict(x_test)
```
